Assume we have a wave time series within interval $[0,T]$ (where $T$ is time, say few hours) as follows

**What is probability of waves crests exceed a certain value $h$ ($h\gg 1$)?**

Probability of crest height, $c$, greater than $h$ is written mathematically as $\rm {Pr}[C>h]$:

$$\rm {Pr}[C>h]=\frac{\rm number~ of~ wave~ crest > h}{\rm{ total~ number ~of ~waves}}$$


In above figure, we count waves using zero-upcrossing method (__zero crossing analysis__) in which we first draw the mean line, zero-level (here $x$-axis) and then the signal crosses the zero-level in upward direction - we may shows zero-upcrossing by $T_{02}$). We conduct the same upcrossing analysis and instead of zero-level, we use $h$-level. This method is called __$h$-upcrossing__


In this example, total number of waves is $7$ and number of crests greater $h$ (dashed black line) is $3$. Therefore


$$\rm {Pr}[C>h]=\frac{3}{7}$$
